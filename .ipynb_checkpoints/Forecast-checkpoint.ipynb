```python
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
import os
import xarray as xr
```

```python
%config InlineBackend.figure_format = 'retina'
RANDOM_SEED = 8927
np.random.seed(RANDOM_SEED)
az.style.use("arviz-darkgrid")
```

```python
def get_dates(today = dt.datetime.today()):
    '''
    Dado o dia de hoje, computa todas as datas de quinze em quinze dias, começando por amanhã
    e terminando em 360 dias. Retorna as datas numa lista em formato %d%m%Y
    '''
    list_days = [1] + [i*15 for i in range(1, 25)]
    dates = [(today + dt.timedelta(days=day)) for day in list_days]
    dates_mdY = [date.strftime('%m/%d/%Y') for date in dates]
    dates_dmY = [date.strftime('%d/%m/%Y') for date in dates]
    dates_hifen = [date.strftime('%d-%m-%Y') for date in dates]
    
    return dates_mdY, dates_dmY, dates_hifen
```

```python
test_df = pd.DataFrame(columns = ['Companhia', 
                                  'Data de Coleta', 
                                  'Dias Após Coleta', 
                                  'Preço Mínimo',
                                  'Média dos Preços'])

for folder in os.listdir('./Azul')[1:]:
    #print(folder)
    tmp_df = pd.DataFrame()
    _, _, tmp_dates = get_dates(dt.datetime.strptime(folder, '%d-%m-%Y'))
    list_days = [1] + [i*15 for i in range(1, 25)]
    i = 0
    for d in tmp_dates:
        #print(d)
        try:
            date_df = pd.read_pickle("./Azul/" + folder + f"/{d}.pkl")
            prices = date_df["Preço (R$)"]          
        except:
            prices = np.nan
        
        try:
            min_price = prices.min()
            avg_price = prices.mean()
        except:
            min_price = np.nan
            avg_price = np.nan
            
        new_row = {'Companhia': 'Azul', 
                'Data de Coleta': folder, 
                'Dias Após Coleta': list_days[i], 
                'Preço Mínimo': [min_price],
                'Média dos Preços': round(avg_price, 2) }
        test_df = pd.concat([test_df, pd.DataFrame.from_dict(new_row)])
        
        i += 1
test_df.reset_index(drop=True, inplace=True)
```

```python
test_df = test_df.dropna()
test_df = test_df.sort_values(by=['Dias Após Coleta'])
test_df['Dias Após Coleta'] = pd.to_numeric(test_df['Dias Após Coleta'])
```

```python
test_df['Dias Após Coleta']
```

```python
x = pd.to_numeric(test_df['Dias Após Coleta'])
y = test_df['Média dos Preços']
```

```python
with pm.Model() as model:  # model specifications in PyMC are wrapped in a with-statement
    # Define priors
    sigma = pm.HalfCauchy("sigma", beta=10)
    intercept = pm.Normal("Intercept", 1500, sigma=500)
    slope = pm.Normal("slope", 10, sigma=20)

    # Define likelihood
    likelihood = pm.Normal("y", mu=intercept + slope * x, sigma=sigma, observed=y)

    # Inference!
    # draw 3000 posterior samples using NUTS sampling
    idata = pm.sample(3000)
```

```python
az.plot_trace(idata, figsize=(10, 7));
```

```python
idata.posterior["y_model"] = idata.posterior["Intercept"] + idata.posterior["slope"] * xr.DataArray(x)
```

```python
_, ax = plt.subplots(figsize=(20, 10))
az.plot_lm(idata=idata, y="y", num_samples=100, axes=ax, y_model="y_model")
ax.set_title("Posterior predictive regression lines")
ax.set_xlabel("x");
```

# Fazer Teste com Splines

```python
test_df.plot.scatter('Dias Após Coleta', 'Média dos Preços')
```

```python
num_knots = 25
knot_list = np.quantile(test_df['Dias Após Coleta'], np.linspace(0, 1, num_knots))
knot_list
```

```python
test_df.plot.scatter(
    'Dias Após Coleta', 'Média dos Preços', color="cornflowerblue", s=10
)
for knot in knot_list:
    plt.gca().axvline(knot, color="grey", alpha=0.4);
```

```python
from patsy import dmatrix

B = dmatrix(
    "bs(days, knots=knots, degree=3, include_intercept=True) - 1",
    {"days": test_df['Dias Após Coleta'], "knots": knot_list[1:-1]},
)
B
```

```python
spline_df = (
    pd.DataFrame(B)
    .assign(days=test_df['Dias Após Coleta'])
    .melt('days', var_name="spline_i", value_name="value")
)

color = plt.cm.magma(np.linspace(0, 0.80, len(spline_df.spline_i.unique())))

fig = plt.figure()
for i, c in enumerate(color):
    subset = spline_df.query(f"spline_i == {i}")
    subset.plot('days', "value", c=c, ax=plt.gca(), label=i)
plt.legend(title="Spline Index", loc="upper center", fontsize=8, ncol=6);
```

```python
COORDS = {"splines": np.arange(B.shape[1])}
with pm.Model(coords=COORDS) as spline_model:
    a = pm.Normal("a", 1500, 500)
    w = pm.Normal("w", mu=1000, sigma=250, size=B.shape[1], dims="splines")
    mu = pm.Deterministic("mu", a + pm.math.dot(np.asarray(B, order="F"), w.T))
    sigma = pm.HalfCauchy("sigma", beta=20)
    D = pm.Normal("D", mu=mu, sigma=sigma, observed=test_df['Média dos Preços'])
```

```python
with spline_model:
    idata = pm.sample_prior_predictive()
    idata.extend(pm.sample(draws=2000, tune=1000, random_seed=RANDOM_SEED, chains=4))
    pm.sample_posterior_predictive(idata, extend_inferencedata=True)
```

```python
az.summary(idata, var_names=["a", "w", "sigma"])
```

```python
az.plot_trace(idata, var_names=["a", "w", "sigma"]);
```

```python
az.plot_forest(idata, var_names=["w"], combined=False, r_hat=True);
```

```python
post_pred = az.summary(idata, var_names=["mu"]).reset_index(drop=True)
test_df_post = test_df.copy().reset_index(drop=True)
test_df_post["pred_mean"] = post_pred["mean"]
test_df_post["pred_hdi_lower"] = post_pred["hdi_3%"]
test_df_post["pred_hdi_upper"] = post_pred["hdi_97%"]
test_df_post['Dias Após Coleta'] = pd.to_numeric(test_df_post['Dias Após Coleta'])
```

```python
test_df.plot.scatter(
    'Dias Após Coleta', 'Média dos Preços', color="cornflowerblue", s=10
)
for knot in knot_list:
    plt.gca().axvline(knot, color="grey", alpha=0.4);

test_df_post.plot("Dias Após Coleta", "pred_mean", ax=plt.gca(), lw=3, color="firebrick")
plt.fill_between(
    test_df_post['Dias Após Coleta'],
    test_df_post.pred_hdi_lower,
    test_df_post.pred_hdi_upper,
    color="firebrick",
    alpha=0.4,
);
```
